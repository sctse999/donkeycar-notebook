# Clean all tubs

```python
import os
from PIL import Image
import shutil

def create_black_image(path, width=160, height=120):
    """Create a black image of specific dimensions."""
    image = Image.new('RGB', (width, height), color='black')
    image.save(path, 'JPEG')

def scan_and_replace_empty_images(root_dir, black_image_path):
    """Scan all subfolders for empty jpg files and replace them."""
    for dirpath, dirnames, filenames in os.walk(root_dir):
        count = 0
        for filename in filenames:
            if filename.endswith('.jpg'):
                file_path = os.path.join(dirpath, filename)
                if os.path.getsize(file_path) == 0:
                    count += 1
                    # Replace the empty image with the black image
                    print(f"Replacing {file_path}") 
                    shutil.copy(black_image_path, file_path)
                    
        if count > 0:
            print(f"Replaced {count} empty image(s) in {dirpath}")

# Path where the black image will be stored
black_image_path = '/tmp/black.jpg'
# Create the black image
create_black_image(black_image_path)

# Path to the directory to scan
root_dir = '/home/pi/mycar'
# Scan directories and replace empty images
scan_and_replace_empty_images(root_dir, black_image_path)
```
